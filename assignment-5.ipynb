```python
import pandas as pd

# Load the dataset
df = pd.read_csv("D:/web dev/internship/Data for repository.csv")

# Movie with the highest budget
highest_budget_movie = df.loc[df['Budget(INR)'].idxmax()]

# Movie with the lowest revenue
lowest_revenue_movie = df.loc[df['Revenue(INR)'].idxmin()]

# Display results
print("Movie with Highest Budget:")
print("Movie Name:", highest_budget_movie['Movie_Name'])
print("Budget:", highest_budget_movie['Budget(INR)'])

print("\nMovie with Lowest Revenue:")
print("Movie Name:", lowest_revenue_movie['Movie_Name'])
print("Revenue:", lowest_revenue_movie['Revenue(INR)'])
```
